**11 - Grayscale Image**


---



*   np.arrange(n) to create array of size n in x-axis
*   we can also use np.array([],[]...) to create array.
*   to add new axis in array we use np.newaxis
*   np.repeat for multiply the array here 10 means 10 times and axis=0 means vertically (y-axis)
*   plt.imshow(nameOfTheImage, cmap='gray', vmin="0", vmap="255") will create graph where vmin and vmap represent the value of max and min colour. The bydefault value of vmin and vmax are min size of the array and min size of the array respectively. 



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.arange(32)         # 32 in x-axis
# img.shape

# np.newaxis to add new axis
img = img[np.newaxis,:]
# img.shape

# np.repeat for multiply the array here 10 means 10 times and axis=0 means vertically (y-axis)
img = np.repeat(img, 10, axis=0)
# img.shape
# img

plt.imshow(img, cmap='gray')
# plt.imshow(img, cmap='gray', vmin="0", vmax="255")
plt.show()
```

**28 - RGB to HSV algorithm implementation using python**


---



```python
def rgbTOhsv(r, g, b, scaleFactor):
  r, g, b = r/255.0, g/255.0, b/255.0
  cmax = max(r, g, b)
  cmin = min(r, g, b)
  diff = cmax-cmin
  if cmax==cmin:
    h = 0
  elif cmax==r:
    h = (60*((g-b)/diff)+0)%360
  elif cmax==g:
    h = (60*((b-r)/diff)+120)%360
  elif cmax==b:
    h = (60*((r-g)/diff)+240)%360

  if h<0:
    h += 360
  if cmax==0:
    s = 0
  else:
    s = (diff/cmax)*scaleFactor
  v = cmax*scaleFactor
  return h, s, v
```

```python
print(rgbTOhsv(100, 200, 50, 1))
```

**31 - Red Rose extraction or Segmentation using HSV Python**


---



*   So basically, its all about color ranges. For colour ranges we can google it.
* sv2.bitwise_and(...) to merge to image whereas mask part is optional.



```python
lowerRed = np.array([0,120,70])
upperRed = np.array([8,255,255])
mask1 = cv2.inRange(hsvImg, lowerRed, upperRed)

plt.imshow(mask1)
```

```python
lowerRed = np.array([175,120,70])
upperRed = np.array([180,255,255])
mask2 = cv2.inRange(hsvImg, lowerRed, upperRed)

plt.imshow(mask2)
```

```python
mask = mask1 | mask2     # bitwise OR operator
```

```python
res = cv2.bitwise_and(rgbImg, rgbImg, mask=mask)
```

```python
plt.figure(1)

plt.subplot(121)
plt.imshow(rgbImg)

plt.subplot(122)
plt.imshow(res)

plt.show()
```
